### Part 1: Quantitative Analysis of Industry Sector
#### A. Industry Sector Selection and Data Filtering 

```python

# 1 . load groups
import pandas as pd
major_groups = pd.read_csv("data/major_groups.csv")
groups_chosen = major_groups.iloc[30:34]

# 2 . filter by groups
public_firms = pd.read_csv("data/public_firms.csv")
df = public_firms.loc[(public_firms['sic']//100).isin([23])]

# 3 . stats
print("there are {} unique years in df".format(len(df.fyear.unique())))
print("there are {} unique companies in df".format(len(df.conm.unique())))
print("there are {} companies in df have 27 years' records".format(
    len([len(i) for i in df.groupby("conm").fyear.unique() if len(i) > 26])))

# df[['conm','fyear']].drop_duplicates().conm.value_counts()>26
```

#### B. Preliminary Analysis

```python
# 1 top 10 stocks in 2020
df.loc[public_firms['fyear'] == 2020 ].sort_values('prcc_c',ascending=False)[:10][['conm','prcc_c']]
```

```python
# 2: top 10 sales of all time
df.sort_values("sale",ascending=False).drop_duplicates(subset='conm')[:10][['conm','sale','fyear']]
```

```python
# 3: top 10 locations
df[['location','conm']].drop_duplicates().location.value_counts()[:10]
df
```
